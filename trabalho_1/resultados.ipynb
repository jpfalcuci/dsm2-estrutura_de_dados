```python
tamanhos = ['10k', '25k', '50k', '100k']
df = pd.DataFrame(columns=['Tempo Gasto', 'Pico de memória (MB)', 'Comparações', 'Trocas', 'Passadas', 'Divisões', 'Junções', 'Algoritmo', 'Tamanho'])

for tamanho in tamanhos:    
    data = [bubble_sort[tamanho], selection_sort[tamanho], merge_sort[tamanho], merge_sort_iter[tamanho], quick_sort[tamanho], quick_sort_iter[tamanho]]
    algoritmos = ['Bubble Sort', 'Selection Sort', 'Merge Sort', 'Merge Sort Iter', 'Quick Sort', 'Quick Sort Iter']
    df_temp = pd.DataFrame(data)
    df_temp['Tempo Gasto'] = pd.to_timedelta(df_temp['Tempo Gasto'])
    df_temp['Tempo (segundos)'] = df_temp['Tempo Gasto'].dt.total_seconds()
    df_temp['Algoritmo'] = algoritmos
    df_temp['Tamanho'] = tamanho

    df = pd.concat([df, df_temp])

df['Pico de memória (MB)'] = df['Pico de memória (MB)'].astype(float)
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
df
```

```python
sns.set_theme(style="darkgrid")
sns.relplot(x='Tamanho', y='Pico de memória (MB)', kind='line', data=df, errorbar=None, hue='Algoritmo', height=5, aspect=1.8).set(title='Consumo de memória x Tamanho da lista - Linhas')
```
